```python
import seaborn as sns
import numpy as np
```

```python
diamon = sns.load_dataset("diamonds")
```

```python
df = diamon.copy()
df.head(5)
```

```python
##### ordinal tanimlama
from pandas.api.types import CategoricalDtype
df.cut = df.cut.astype(CategoricalDtype( ordered=True))
```

```python
cut_catogory = ["Fair","Good","Very Good","Premium","Ideal"]
df.cut = df.cut.astype(CategoricalDtype(categories=cut_catogory, ordered=True))
```
